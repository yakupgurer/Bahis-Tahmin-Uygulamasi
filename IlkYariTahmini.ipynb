```python
# Gerekli kütüphaneleri import ediyoruz
import pandas as pd
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

veri_kumesi = pd.read_csv('D:/YENI_CEKILEN_VERILER.csv')
veri_kumesi = veri_kumesi.fillna(0)
X = veri_kumesi.drop(['iy','iyev','iydep','msev','msdep'], axis=1)  # Girdi verisi (bağımsız değişkenler)
y = veri_kumesi['iy']  # Hedef veri (bağımlı değişken)

###XGBoost Model İçin



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)

# XGBoost sınıflandırıcı modelini oluşturuyoruz
xgb_model = XGBClassifier(random_state=8)

# Modeli eğitiyoruz
xgb_model.fit(X_train, y_train)


tahmin_edilecek_mac=pd.read_csv("C:/tahmin2.csv")

y_pred=xgb_model.predict(tahmin_edilecek_mac)
   

print("XGBoost Tahmin :",y_pred)

### SVM Model için

# Girdi ve hedef veri kümesini eğitim ve test verilerine ayırıyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=61)

# SVM sınıflandırıcı modelini oluşturuyoruz
svm_model = SVC(kernel='rbf', C=450.0)  # Lineer çekirdek kullanarak C=1.0 olarak belirledik

# Modeli eğitiyoruz
svm_model.fit(X_train, y_train)


tahmin_edilecek_mac=pd.read_csv("C:/tahmin2.csv")

y_pred=svm_model.predict(tahmin_edilecek_mac)
print("SVM Tahmin: ", y_pred)

### KNN İçin

# Gerekli kütüphaneleri import ediyoruz
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Girdi ve hedef veri kümesini eğitim ve test verilerine ayırıyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=29)

# KNN sınıflandırıcı modelini oluşturuyoruz
knn_model = KNeighborsClassifier(n_neighbors=35)  # K=5 olarak belirledik

# Modeli eğitiyoruz
knn_model.fit(X_train, y_train)


tahmin_edilecek_mac=pd.read_csv("C:/tahmin2.csv")

y_pred=svm_model.predict(tahmin_edilecek_mac)
print("KNN Tahmin: ", y_pred)

#######  Naive bayes - Gaussian NB

# Gerekli kütüphaneleri import ediyoruz
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Girdi ve hedef veri kümesini eğitim ve test verilerine ayırıyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

# Naive Bayes sınıflandırıcı modelini oluşturuyoruz
nb_model = GaussianNB()

# Modeli eğitiyoruz
nb_model.fit(X_train, y_train)

tahmin_edilecek_mac=pd.read_csv("C:/tahmin2.csv")

y_pred=svm_model.predict(tahmin_edilecek_mac)
print("Naive Bayes-Gaussian NB Tahmin: ", y_pred)


#### Naive Bayes - Multinomial NB


# Gerekli kütüphaneleri import ediyoruz
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB


# Girdi ve hedef veri kümesini eğitim ve test verilerine ayırıyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)

# Multinomial Naive Bayes modelini oluşturuyoruz
multinomial_naive_bayes_model = MultinomialNB()

# Modeli eğitiyoruz
multinomial_naive_bayes_model.fit(X_train, y_train)

tahmin_edilecek_mac=pd.read_csv("C:/tahmin2.csv")

y_pred=svm_model.predict(tahmin_edilecek_mac)
print("Naive Bayes-Multinomial NB Tahmin: ", y_pred)


##### Karar agaci


# Gerekli kütüphaneleri import ediyoruz
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Girdi ve hedef veri kümesini eğitim ve test verilerine ayırıyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=14)

# Karar ağacı modelini oluşturuyoruz
karar_agaci_modeli = DecisionTreeClassifier()

# Modeli eğitiyoruz
karar_agaci_modeli.fit(X_train, y_train)

tahmin_edilecek_mac=pd.read_csv("C:/tahmin2.csv")

y_pred=svm_model.predict(tahmin_edilecek_mac)
print("Karar Agaclari Tahmin: ", y_pred)


### Random Forest

import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)

rf_model = RandomForestClassifier(n_estimators=420, random_state=8)  # 100 ağaçtan oluşan bir model
rf_model.fit(X_train, y_train)


tahmin_edilecek_mac=pd.read_csv("C:/tahmin2.csv")

y_pred=rf_model.predict(tahmin_edilecek_mac)
print("Random Forest Tahmin: ", y_pred)
```
